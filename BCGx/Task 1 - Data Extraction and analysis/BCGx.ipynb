```python
import pandas as pd
```

# Manually Extracted the dataset from SEC's EDGAR's databse

```python
df = pd.read_csv('final_report.csv')
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
```

```python
df
```

# Imputing the null values with 0

```python
df.fillna(0, inplace=True)
```

```python
df
```

# Calculating Year-by-Year (YOY) growth for Total Assests/Liabilities and Cash flow 

```python
df['Assets Growth(%)'] = df.groupby('Company')['Total Assets'].pct_change()*100
df['Liabilities Growth(%)'] = df.groupby('Company')['Total Liabilities'].pct_change()*100
df['Cash Flow Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change()*100
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

# Exporting the Analyzed dataset

```python
df.to_csv('analyzed.csv')
```

# Summarizing 

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth(%)': 'mean',
    'Liabilities Growth(%)': 'mean',
    'Cash Flow Growth(%)' :'mean'  
}).reset_index()
```

```python
print('Year-by-Year Average growth rates(%): ->')
summary
```

# Eporting the summarized report

```python
summary.to_csv('Summarized_report.csv')
```
